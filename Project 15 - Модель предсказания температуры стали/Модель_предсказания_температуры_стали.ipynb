### Вычислим мощность и энергию, затраченную на нагрев.

*Вернёмся к выбросу в столбце 'Реактивная мощность'. Посмотрим партию.*

```python
data_arc[(data_arc['key'] == 2116)]
```

*Так как отрицательное значение не последнее по партии, то удалим его.*

```python
data_arc = data_arc.loc[data_arc['Реактивная мощность']>=-100]
```

```python
# проверим
data_arc[(data_arc['key'] == 2116)]
```

```python
# посмотрим значение корреляции между 'Активная мощность' и 'Реактивная мощность'
data_arc['Активная мощность'].corr(data_arc['Реактивная мощность'])
```

*На лицо сильная зависимось значений. Оставлять два признака нельзя, так как качество будущей модели будет страдать. Приведем два этих значения к одному. Вычислим полную мощность.*

$$
S=(P^{2}+R^{2})^{0.5}
$$
где P - активная мощность,\
R - реактивня мощность.

```python
data_arc['Полная мощность'] = (data_arc['Активная мощность']**2 + data_arc['Реактивная мощность']**2)**0.5
data_arc.head()
```

*Из полученного значения полной мощности можем получить значения затраченной энергии на нагрев. Для этого вычислим время нагрева и умножим на мощность.*

$$
Q=S  t
$$

где t - время нагрева

```python
# переведём значения в столбцах формат datetime
data_arc['Начало нагрева дугой'] = pd.to_datetime(data_arc['Начало нагрева дугой'])
data_arc['Конец нагрева дугой'] = pd.to_datetime(data_arc['Конец нагрева дугой'])
```

```python
# вычислим дельту
data_arc['Время нагрева'] = data_arc['Конец нагрева дугой'] - data_arc['Начало нагрева дугой']
data_arc['Время нагрева'] = data_arc['Время нагрева'].apply(lambda x: x.seconds) # переведём в секунды
```

```python
data_arc['Затраченная энергия'] = data_arc['Полная мощность'] * 1000 * data_arc['Время нагрева'] / 3600
```

### Объединяем данные в один датасет

*Для объединения данных в один датасет нам нужно сгруппировать данные по key.*

*Выполним группировку по key в data_arc и посчитаем суммарную энергию, затраченную для одной партии.*

```python
data_arc_key = data_arc.groupby('key').agg({'Затраченная энергия': 'sum'}).reset_index()
data_arc_key.head()
```

*Для сыпучих, проволочных материалов и газе группировка по key не требуется*

*Для данных по температуре потребуется группировка по key.\
Также при анализе данных мы видели, что для одной партии делается несколько замеров.\
Отсортируем данные о температуре по key.*

```python
data_temp = data_temp.sort_values(by=['key', 'Время замера'])
data_temp.head(10)
```

*Теперь сгруппируем данные по key, также посчитаем количество замеров в партии и выведем первое и последнее измерение температуры.*

```python
data_temp_key = data_temp.groupby('key').agg({'Температура': ['count', 'first', 'last']}).reset_index()
data_temp_key.tail()
```

```python
# зададим корректное название столбцов
data_temp_key.columns = ['key', 'count',
              'start_temp', 'final_temp']
```

*Если в партии всего один замер температуры, то данных недостаточно, поэтому удалим такие строки.*

```python
data_temp_key = data_temp_key[data_temp_key['count'] > 1]
data_temp_key.head()
```

```python
data_temp_key
```

*Так как информация о температуре является основной для нашей модели, то к данным по температуре, сгруппированным по key, будем добавлять остальные данные.*

*Для объединения нам потребуются датафреймы data_arc_key, data_temp_key, data_bulk, data_gas и data_wire.*

## Обучение моделей

### Линейная регрессия

```python
%%time
model_lr = LinearRegression()

parameters = {'fit_intercept':[True,False], 'normalize':[True,False]}
grid = RandomizedSearchCV(model_lr, parameters, cv=5, random_state=240423,
                          scoring='neg_mean_absolute_error', verbose=5)
grid.fit(features_train, target_train)

MAE_lr_train=-grid.best_score_

print('Наилучшие параметры для LinearRegression:', grid.best_params_)
print('MAE LinearRegression на обучающей выборке:', MAE_lr_train)
```

### RandomForestRegressor

```python
%%time
model_rfr = RandomForestRegressor(random_state=240423)

params = {'n_estimators': range(100, 500, 50),
                 'max_depth': range(1,100,2)}

forest_grid = RandomizedSearchCV(model_rfr, params, cv=5,
                                 scoring='neg_mean_absolute_error', verbose=5, n_jobs=-1)
forest_grid.fit(features_train, target_train)

MAE_rfr_train = -forest_grid.best_score_

print('MAE RandomForestRegressor на обучающей выборке:', MAE_rfr_train)

print('Наилучшие параметры для RandomForestRegressor:', forest_grid.best_params_)
```

### Catboost

```python
%%time
model_cbr = CatBoostRegressor(loss_function="MAE", random_state=240423)

#grid = {'n_estimators': [100, 250, 500],
#        'depth': [4, 10]}
grid = {
    'learning_rate': [x/100 for x in range(1,10)],
    'iterations': range(50, 300)
}

cbr_grid = RandomizedSearchCV(model_cbr, grid,
                         cv=5, verbose=50, scoring='neg_mean_absolute_error')

cbr_grid.fit(features_train, target_train)

MAE_cbr_train = -cbr_grid.best_score_
print('MAE для CatBoostRegressor на обучающей выборке:', MAE_cbr_train)

print('Наилучшие параметры для CatBoostRegressor:', cbr_grid.best_params_)
```

### LightGBMRegressor

```python
%%time
model_gbm = LGBMRegressor(random_state=240423)

params = {'n_estimators': [x for x in range(50,300)],
              'num_leaves': [x for x in range(10,50)],
              'learning_rate':[x/100 for x in range(1, 10)]}

g_grid = RandomizedSearchCV(model_gbm, params,
                         cv=5, verbose=5, scoring='neg_mean_absolute_error')
g_grid.fit(features_train, target_train)
MAE_gbm_train = -g_grid.best_score_
print('MAE для LGBMRegressor на обучающей выборке', MAE_gbm_train)

print('Наилучшие параметры для LGBMRegressor:', g_grid.best_params_)
```

```python
list_train = (MAE_lr_train, MAE_rfr_train, MAE_cbr_train, MAE_gbm_train)

df_mae=pd.DataFrame(list_train,
                           index=['LinearRegression',
                                  'RandomForestRegressor',
                                  'CatBoostRegressor',
                                  'LGBMRegressor'],
                           columns = ['MAE'])
df_mae
```
